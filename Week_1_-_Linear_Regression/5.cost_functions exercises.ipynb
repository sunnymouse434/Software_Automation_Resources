If you did the step above correctly, your error was `-200`.

To fix this, we **square the error** (multiply the number by itself). This automatically turns all negative numbers into positive numbers.

````python
### SAMPLE CODE: How to square a number in Python ###

sample_error = -20

# In Python, we use two asterisks (**) to square a number.
sample_squared = sample_error ** 2

print(f"[Sample] The squared error is: {sample_squared}")
````

### Step 5: Put It All Together!

You did it! You just learned all 4 pieces of the Cost Function equation:
$$ J(w, b) = \frac{1}{2n} \sum_{i=1}^{n} \left( f_{w,b}(x^{(i)}) - y^{(i)} \right)^2 $$

Now, we are going to write it into a proper Python Function. 
Study the **Sample Code** below. It calculates a simpler "Absolute Error". 
Then, scroll down to the **Exercise Code** and write the real "Cost Function" by replacing the missing pieces.

````python
### SAMPLE CODE: Absolute Error Function ###
# Study this structure! You will copy this logic in the next cell.

def compute_absolute_error(x, y, w, b):
    n = x.shape[0]  # Total number of data points
    cost_sum = 0    # Starting total
    
    for i in range(n):
        # 1. Make the prediction
        prediction = w * x[i] + b  
        
        # 2. Find the error (In this sample, we just use absolute value 'abs()')
        error = abs(prediction - y[i]) 
        
        # 3. Add to the running total
        cost_sum = cost_sum + error
        
    # 4. Divide to get the average
    final_cost = cost_sum / n
    return final_cost
````
