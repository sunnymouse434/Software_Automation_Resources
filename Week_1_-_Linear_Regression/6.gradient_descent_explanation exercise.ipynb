# Gradient Descent: See It, Then Try It

Welcome back! You already know how to check if a line is a "Good Fit" using the Cost Function. 
But how does the computer actually *find* the best line on its own, without a human manually tweaking the dials?

It uses an algorithm called **Gradient Descent**. 

### The Mountain Analogy
Imagine you are blindfolded at the top of a mountain, and your goal is to get to the very bottom of the valley (zero Cost/Error). 
Because you are blindfolded, you can only feel the slope of the ground right beneath your feet.
1. You feel the ground. If it tilts down to the left, you take a step left.
2. You take a step.
3. You feel the ground again, and repeat until the ground is flat!

Let's build this step-by-step.

### Step 1: Feeling the Slope (The Derivative)

In math, the "slope of the ground beneath your feet" is called a **Derivative** (or Gradient). 
It tells us which way is up, and which way is down.

* If the slope is **Positive** (pointing up), we need to go backward (decrease our weight) to go down.
* If the slope is **Negative** (pointing down), we need to go forward (increase our weight) to keep going down.

````python
### SAMPLE CODE: Which way is down? ###

sample_w = 5
sample_slope = 2.0  # The slope is positive (going up!)

if sample_slope > 0:
    print("Slope is positive. We must decrease w to go down the mountain.")
    sample_w = sample_w - 1
else:
    print("Slope is negative. We must increase w to go down the mountain.")
    sample_w = sample_w + 1
    
print(f"[Sample] New w is: {sample_w}")
````

### Step 2: Taking a Step (The Learning Rate)

Instead of just taking a random step of "1", we want to take a step based on exactly how steep the mountain is. 
We multiply the slope by a tiny fraction called the **Learning Rate** (represented by the symbol $\alpha$, or "alpha"). 

* If $\alpha$ is too small, the hiker takes baby steps and it takes forever to reach the bottom.
* If $\alpha$ is too big, the hiker takes giant leaps and might accidentally jump entirely over the valley!

````python
### SAMPLE CODE: Taking a controlled step ###

w_old = 10
slope = 4.0
alpha = 0.1  # Our learning rate (step size)

# Update Rule: New Weight = Old Weight - (Step Size * Slope)
w_new = w_old - (alpha * slope)

print(f"[Sample] We took a step! w changed from {w_old} to {w_new}")
````

### Step 3: Calculating the Gradient (The Math)

How do we actually calculate the slope? We use our old friend, the Error (`Prediction - Actual`).

* The slope for the Elevator (`b`) is just the Error.
* The slope for the Seesaw (`w`) is the Error *multiplied by the Feature (`x`)*.

Study how the sample code calculates this for *one* data point, then try calculating it for a whole list!

````python
### SAMPLE CODE: Slope of one point ###

x_point = 2
y_actual = 10
prediction = 14

error = prediction - y_actual

# Gradient for b (dj_db) is just the error
dj_db_point = error

# Gradient for w (dj_dw) is error multiplied by x
dj_dw_point = error * x_point

print(f"[Sample] Gradient for b: {dj_db_point}, Gradient for w: {dj_dw_point}")
````

````python
### YOUR TURN: Exercise 3 ###

def compute_gradient(x, y, w, b): 
    n = x.shape[0]    
    dj_dw = 0
    dj_db = 0
    
    for i in range(n):  
        # 1. Make the prediction
        prediction = w * x[i] + b 
        
        # 2. Find the error
        error = prediction - y[i] 
        
        # 3. TODO: Find the gradient for w (error times x[i])
        dj_dw_i = 0
        
        # 4. TODO: Find the gradient for b (just the error)
        dj_db_i = 0
        
        # Add them to the running total
        dj_db = dj_db + dj_db_i
        dj_dw = dj_dw + dj_dw_i 
        
    # Divide by n to get the average slope over all points
    return (dj_dw / n), (dj_db / n)

print("[Exercise 3] Gradient function saved!")
````

--- 
### Step 4: The Iteration Loop

A hiker doesn't just take one step and teleport to the bottom. They take hundreds of steps! 
We need to loop our Update Rule hundreds of times (called `iterations`) until we reach the bottom.

````python
### SAMPLE CODE: The Gradient Descent Loop ###

def gradient_descent(x, y, w_in, b_in, alpha, num_iters):
    w = w_in
    b = b_in
    
    # Loop to take steps down the mountain
    for i in range(num_iters):
        # 1. Feel the slope (Calculate Gradients)
        dj_dw, dj_db = compute_gradient(x, y, w, b)     

        # 2. Take a step (Update Rule)
        b = b - (alpha * dj_db)                            
        w = w - (alpha * dj_dw)  
        
    return w, b

print("[Sample] Gradient Descent function saved!")
````
