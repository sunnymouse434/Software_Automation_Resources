### Your Toolkit (The Given Functions)

You already know how to write the Cost, Gradient, and Loop functions. To save you time, we have written them for you below so you can focus purely on optimizing them! 

**Run the cell below to load your toolkit.**

````python
### RUN ME: Your Pre-Built Toolkit ###

def compute_cost(x, y, w, b): 
    n = x.shape[0] 
    cost_sum = 0 
    for i in range(n): 
        f_wb = w * x[i] + b   
        cost = (f_wb - y[i]) ** 2  
        cost_sum += cost  
    return (1 / (2 * n)) * cost_sum  

def compute_gradient(x, y, w, b): 
    n = x.shape[0]    
    dj_dw = 0
    dj_db = 0
    for i in range(n):  
        f_wb = w * x[i] + b 
        dj_dw += (f_wb - y[i]) * x[i] 
        dj_db += (f_wb - y[i]) 
    return (dj_dw / n), (dj_db / n)

def gradient_descent(x, y, w_in, b_in, alpha, num_iters): 
    J_history = [] # This list will save our 'cost' at every single step!
    w = w_in
    b = b_in
    for i in range(num_iters):
        dj_dw, dj_db = compute_gradient(x, y, w , b)     
        b = b - alpha * dj_db                            
        w = w - alpha * dj_dw                            
        J_history.append(compute_cost(x, y, w, b))
    return w, b, J_history

print("Toolkit loaded! You are ready to go.")
````

--- 
### Step 1: The Danger of Big Steps (Divergence)

The Learning Rate ($\alpha$) controls how big of a step the hiker takes. 
* If it is small, the hiker safely walks down to the bottom.
* If it is too big, the hiker overshoots the valley, steps way up the other side of the mountain, and bounces back and forth into space! We call this **Divergence**.

````python
### SAMPLE CODE: A safe learning rate ###
x_train = np.array([1.0, 2.0])   
y_train = np.array([300.0, 500.0]) 

safe_alpha = 0.01
iters = 10

w_final, b_final, J_history = gradient_descent(x_train, y_train, 0, 0, safe_alpha, iters)

print("[Sample] Look at the Cost History. It is safely going DOWN towards zero:")
for i, cost in enumerate(J_history[:5]):
    print(f"Step {i+1}: Cost is {cost:.2f}")
````

### Step 2: Leveling the Playing Field (Data Normalization)

Look at our training data: `x` is tiny (1.0, 2.0) but `y` is huge (300.0, 500.0). 
When data is drastically different in size, the "mountain" of our Cost Function becomes stretched out and warped, making it very hard for the hiker to find the bottom. 

**The Solution:** We "Normalize" or "Scale" the data so everything falls between 0 and 1. 
The math for this is simple: 

$X_{norm} = \frac{X - Min}{Max - Min}$

````python
### SAMPLE CODE: How to Normalize Data ###

sample_data = np.array([10, 20, 30, 40, 50])

# Find the minimum and maximum
data_min = sample_data.min()
data_max = sample_data.max()

# Apply the formula: (Data - Min) / (Max - Min)
normalized_data = (sample_data - data_min) / (data_max - data_min)

print(f"[Sample] Original Data: {sample_data}")
print(f"[Sample] Normalized Data: {normalized_data}")
````

### Bonus Visualizer: The False Bottom (Multiple Minima)

Up until now, our cost functions have looked like a perfect bowl. There is only one bottom (the **Global Minimum**). 

But as your models get more complex, the cost function will look like a bumpy landscape with many dips and valleys. If your hiker stops in a shallow valley (a **Local Minimum**), they might think they hit the absolute bottom, when really, a deeper valley exists over the next ridge!

Run the code below to see a 3D visualizer of a complex Cost Function.

````python
### SAMPLE VISUALIZER: Run me! ###
x = np.linspace(-10, 10, 500)
y = np.linspace(-10, 10, 500)
X, Y = np.meshgrid(x, y)
Z = (X**2 + Y**2) * np.sin(X / 2) * np.sin(Y / 2) + 10 * np.exp(-((X - 5)**2 + (Y - 5)**2)) + 8 * np.exp(-((X + 5)**2 + (Y + 5)**2))
fig = plt.figure(figsize=(18, 6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_title("Surface Plot with Multiple Lowest Cost Points")
ax.set_xlabel("Feature w")
ax.set_ylabel("Feature b")
ax.set_zlabel("Cost / Error")
fig.colorbar(surf, shrink=0.5, aspect=10)
plt.show()
````
