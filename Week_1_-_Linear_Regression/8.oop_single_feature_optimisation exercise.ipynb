# Object-Oriented Linear Regression: See It, Then Try It

Welcome to the professional leagues! In the previous lessons, you built Gradient Descent loops completely from scratch to find the perfect weight and bias.

In the real world, Data Scientists don't write those loops from scratch every day. Instead, they use powerful toolkits that do all the heavy math automatically. The most famous toolkit in Python is called **Scikit-Learn** (`sklearn`).

Think of Scikit-Learn as the "Easy Button". Let's learn how to use it!

````python
# Step 0: Run this cell first to load our tools!
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Here are our new professional tools!
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

plt.style.use('ggplot')
print("Tools loaded successfully!")
````

### Your Dataset

In this exercise, you will be predicting the **Latitude** of male insects based on their **Wing Size**. 
Run the cell below to load your data. It will automatically prepare `X_insects` (your features) and `y_insects` (your targets).

````python
### RUN ME: Data Loader ###

try:
    # Try to load the real data file
    insects = pd.read_csv('data/insects.csv', sep='\t')
    insects = insects[insects.sex == 1] # Filter to only male insects
    
    X_insects = insects[['wingsize']] # Feature
    y_insects = insects['latitude']   # Target
    print("Successfully loaded data/insects.csv!")
    
except FileNotFoundError:
    # If the file is missing, generate practice data so the lesson doesn't break!
    print("Could not find data/insects.csv. Generating practice insect data instead!")
    X_insects = pd.DataFrame({'wingsize': np.random.uniform(300, 1000, 100)})
    y_insects = X_insects['wingsize'] * 0.05 + np.random.normal(0, 5, 100) + 10

# Let's look at the first 5 rows
print("\nFirst 5 Wing Sizes:\n", X_insects.head())
print("\nFirst 5 Latitudes:\n", y_insects.head())
````

### Step 1: Building and Training the Machine

To use Scikit-Learn, you follow a simple two-step process:
1. **Instantiate:** Create a blank "Brain" (a `LinearRegression` object).
2. **Fit:** Give the brain your `X` and `y` data so it can run Gradient Descent in the background and learn the pattern.

````python
### SAMPLE CODE: Creating and Fitting a Model ###

X_sample = pd.DataFrame({'feature': [1, 2, 3, 4]})
y_sample = pd.Series([2, 4, 6, 8])

# 1. Create the blank brain
sample_model = LinearRegression()

# 2. Train (fit) the brain on the data
sample_model.fit(X_sample, y_sample)

print("[Sample] The model is successfully trained!")
````

### Step 4: The Automated Scorecard (Evaluation)

Instead of writing the Cost Function loop from scratch, Scikit-Learn comes with built-in scorecards to measure your error. 

* **MSE (Mean Squared Error):** The exact Cost Function you built in the previous lessons! Penalizes big mistakes heavily.
* **MAE (Mean Absolute Error):** Just the average distance from the prediction to the real answer. Easier for humans to understand!

````python
### SAMPLE CODE: Automated Error Checking ###
sample_actual_answers = [5, 10, 15]
sample_model_guesses = [4, 11, 14]

# We just pass the actual answers and the guesses into the functions!
mae = mean_absolute_error(sample_actual_answers, sample_model_guesses)
mse = mean_squared_error(sample_actual_answers, sample_model_guesses)

print(f"[Sample] MAE: {mae}  |  MSE: {mse}")
````
